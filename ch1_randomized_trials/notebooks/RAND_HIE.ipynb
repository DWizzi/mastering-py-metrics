```python
import numpy as np
import pandas as pd

pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
data = pd.read_stata(
    "../data/person_years.dta"
)

data.head()
```

```python
def group_plantypes(plan: int) -> int:
    """Group plans into four types:
        1-Free
        2-Individual Deductible
        3-Cost Sharing (25%/50%)
        4-Catostrophic (Fam Deductible) (95%/100%)

    Args:
        plan (int)

    Returns:
        int: Plantype
    """
    if plan == 24:
        return 1
    if plan in [1, 5]:
        return 2
    if plan in range(9, 24):
        return 3
    if plan in [*range(2,5), *range(6,9)]:
        return 4
    raise ValueError("Invalid input")

data['plantype'] = data['plan'].astype(int).apply(group_plantypes)

data
```
